```python
import numpy as np

# Create a 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D array:", arr1)

# Create a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array:\n", arr2)

# Access array elements
print("Element at index 2 of arr1:", arr1[2])
print("Element at row 1, column 2 of arr2:", arr2[1, 2])

# Array shape and dimensions
print("Shape of arr2:", arr2.shape)
print("Number of dimensions of arr2:", arr2.ndim)

# Array operations
arr3 = np.array([7, 8, 9, 10, 11])
print("Sum of arr1 and arr3:", arr1 + arr3)
print("Product of arr1 and arr3:", arr1 * arr3)

# Array slicing
print("Elements from index 1 to 3 of arr1:", arr1[1:4])
print("First row of arr2:", arr2[0, :])

# Reshaping an array
arr4 = np.arange(12).reshape(3, 4)
print("Reshaped array:\n", arr4)
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame with missing values (NaN)
data = {
'Name': ['John', 'Sara', 'Tom', 'Anna'],
'Age': [28, np.nan, 35, 29],
'Salary': [50000, 45000, np.nan, 52000]
}
df = pd.DataFrame(data)

# Detect Missing Values
print("Detecting Missing Values:")
print(df.isnull())

# Remove Rows with Missing Values
df_dropped = df.dropna()
print("\nDataFrame After Dropping Missing Values (Rows with NaN):")
print(df_dropped)

# Fill Missing Values with a Specific Value (e.g., 0 for Age and Salary)
df_filled = df.fillna(0)
print("\nDataFrame After Filling Missing Values with 0:")
print(df_filled)

# Fill Missing Values with the Mean of the Column
df_filled_mean = df.copy()
df_filled_mean['Age'].fillna(df['Age'].mean(), inplace=True)
df_filled_mean['Salary'].fillna(df['Salary'].mean(), inplace=True)

print("\nDataFrame After Filling Missing Values with Column Mean:")
print(df_filled_mean)
```

```python
import pandas as pd

# Sample DataFrame
data = {
'Name': ['John', 'Sara', 'Tom', 'Anna', 'Peter'],
'Age': [28, 22, 35, 29, 32],
'Salary': [50000, 45000, 60000, 52000, 58000]
}
df = pd.DataFrame(data)

# 1. Sorting the DataFrame by Age in Ascending Order
sorted_df = df.sort_values(by='Age', ascending=True)
print("Sorted DataFrame by Age (Ascending):")
print(sorted_df)

# 2. Filtering Rows where Salary is greater than 50,000
filtered_df = df[df['Salary'] > 50000]
print("\nFiltered DataFrame (Salary > 50,000):")
print(filtered_df)

# 3. Applying a Function to Calculate 10% Salary Increase
def salary_increase(salary):
  return salary * 1.10

df['Increased_Salary'] = df['Salary'].apply(salary_increase)
print("\nDataFrame After Applying 10% Salary Increase:")
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create 1 row, 2 columns subplot
plt.subplot(1, 2, 1)  # First subplot
plt.plot(x, y1, label="sin(x)")
plt.title("Sine Wave")

plt.subplot(1, 2, 2)  # Second subplot
plt.plot(x, y2, label="cos(x)", color='orange')
plt.title("Cosine Wave")

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # Create a 2x2 grid of plots

axes[0, 0].plot(x, y1, label="sin(x)")
axes[0, 0].set_title("Sine Wave")

axes[0, 1].plot(x, y2, label="cos(x)", color='orange')
axes[0, 1].set_title("Cosine Wave")

axes[1, 0].scatter(x, y1, color='green')
axes[1, 0].set_title("Sine Scatter")

axes[1, 1].bar([1, 2, 3, 4], [10, 20, 15, 30], color='purple')
axes[1, 1].set_title("Bar Plot")

plt.tight_layout()  # Adjust layout to avoid overlapping
plt.show()
```

###**Data Aggregation**


```python
import pandas as pd

# Sample data
data = {'Region': ['North', 'South', 'North', 'South', 'East'],
        'Sales': [100, 200, 150, 250, 300]}

df = pd.DataFrame(data)

# Aggregation using sum()
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
data = {'Student': ['Alice', 'Bob', 'Charlie', 'David'],
        'Math': [85, 90, 88, 92],
        'Science': [78, 82, 80, 85]}

df = pd.DataFrame(data)

# Aggregation using mean()
average_scores = df[['Math', 'Science']].mean()
print(average_scores)
```

```python
data = {'Region': ['North', 'South', 'North', 'South', 'East'],
        'Sales': [100, 200, 150, None, 300]}

df = pd.DataFrame(data)

# Aggregation using count()
sales_count = df.groupby('Region')['Sales'].count()
print(sales_count)
```

```python
data = {'Store': ['A', 'B', 'A', 'B', 'C'],
        'Price': [25, 30, 20, 40, 15]}

df = pd.DataFrame(data)

# Aggregation using min()
min_prices = df.groupby('Store')['Price'].min()
print(min_prices)
```

```python
# Sample data
data = {'Department': ['HR', 'Engineering', 'HR', 'Engineering', 'Sales'],
        'Salary': [50000, 75000, 55000, 80000, 60000]}

df = pd.DataFrame(data)

# Aggregation using max()
max_salaries = df.groupby('Department')['Salary'].max()
print(max_salaries)
```

```python
#split-apply-combine

data = {'Region': ['North', 'South', 'North', 'South', 'East', 'East', 'West'],
        'Store': ['A', 'B', 'A', 'B', 'C', 'A', 'B'],
        'Sales': [100, 200, 150, 250, 300, 350, 400]}
df = pd.DataFrame(data)

#split
grouped = df.groupby('Region')
for name, group in grouped:
    print(f"Group: {name}")
    print(group)

#apply
result = grouped["Sales"].sum()

#combine
print(result)
```

```python
import pandas as pd
df = pd.read_csv("/content/daily-minimum-temperatures-in-me.csv")
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Daily minimum temperatures'] = pd.to_numeric(df['Daily minimum temperatures'], errors="coerce")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['Daily minimum temperatures'], color='blue')
plt.title('Daily Minimum Temperatures Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.tight_layout()
plt.show()
```

```python
# prompt: Sample code to Explain the concept of moving window functions in time series analysis.

# Rolling mean
df['Rolling_Mean_7'] = df['Daily minimum temperatures'].rolling(window=7).mean()

# Rolling standard deviation
df['Rolling_Std_7'] = df['Daily minimum temperatures'].rolling(window=7).std()

# Plot the original data and the rolling mean/std
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Daily minimum temperatures'], label='Original Data', color='blue')
plt.plot(df['Date'], df['Rolling_Mean_7'], label='7-Day Rolling Mean', color='red')
# plt.plot(df['Date'], df['Rolling_Std_7'], label='7-Day Rolling Std Dev', color='green')

plt.title('Daily Minimum Temperatures with Rolling Statistics')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.tight_layout()
plt.show()
```
